```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data again and naming as filepath
file_path = 'Google_data.csv'
data = pd.read_csv(file_path)

# Handle 'Size' column conversion from M (MB) and k (KB) to float
data['Size'] = data['Size'].replace('Varies with device', '0')
data['Size'] = data['Size'].str.replace('M', 'e+6').str.replace('k', 'e+3')
data['Size'] = data['Size'].apply(pd.to_numeric, errors='coerce')

# Convert 'Installs' to numerical, remove '+' and ','
data['Installs'] = data['Installs'].str.replace(',', '').str.replace('+', '')
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')

# Convert 'Reviews' to numerical
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')

# Convert 'Price' to numerical, remove '$'
data['Price'] = data['Price'].str.replace('$', '')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Drop rows with missing values in the numerical columns
data = data.dropna(subset=['Size', 'Installs', 'Reviews', 'Price', 'Rating'])

# Compute the correlation matrix for numerical columns
correlation_matrix = data[['Size', 'Installs', 'Reviews', 'Price', 'Rating']].corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
